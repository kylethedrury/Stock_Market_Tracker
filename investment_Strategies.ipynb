```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = np.linspace(0, 45, 46)

def plot_total_gains(first_investment, yearly_investment):
    profits = []
    profits.append(first_investment)

    for i in range(1, 46):
        profits.append(profits[i-1]*1.1 + yearly_investment)

    plt.plot(years, profits, label=f'Yearly investment: {yearly_investment} Inital investment: {first_investment}')

yearly_investments = [1000, 1500, 2000, 2500, 3000, 3500, 4000]

for i in range(7):
    plot_total_gains(25000, year_investments[i])

plt.legend(loc='upper left')
plt.grid()
plt.xlabel('Years')
plt.ylabel('Profits')
plt.show()

for i in range(7):
    plot_total_gains(15000, year_investments[i])

plt.legend(loc='upper left')
plt.grid()
plt.xlabel('Years')
plt.ylabel('Profits')
plt.show()
```
